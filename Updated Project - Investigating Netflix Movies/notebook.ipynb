```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head()
```

```python
# Create a figure and increase the figure size
fig = plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus year
plt.scatter(netflix_df['release_year'] , netflix_df['duration'])

# Create a title
plt.title("Movie Duration by Release Year")

# Show the plot
plt.show()
```

```python
# Filter movies released in the 1990s
movies_90s = netflix_df[(netflix_df['release_year'] >= 1990)]

# Find the most frequent movie duration
duration = movies_90s['duration'].mode()[0]

# Output the result
print(f"The most frequent movie duration in the 1990s is approximately {duration} minutes.")
```

```python
# Filter movies released in the 1990s and of the Action genre
action_movies_90s = movies_90s[(movies_90s['release_year'] >= 1990) & (movies_90s['release_year'] < 2000) & 
                               (movies_90s['type'] == 'Movie') & (movies_90s['genre'] == 'Action')]

# Count the number of short action movies (duration < 90 minutes)
short_movie_count = action_movies_90s[action_movies_90s['duration'] < 90].shape[0]

# Output the result
short_movie_count
```
